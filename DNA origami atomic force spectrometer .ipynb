```python
import pandas as pd
import numpy as np
import openpyxl
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity
from sklearn.model_selection import GridSearchCV
import seaborn
import scipy.stats as stats
from matplotlib.lines import Line2D as line
import math
from scipy.stats import norm
```

```python
data = pd.read_csv(r"C:\Users\lionr\OneDrive\Desktop\Praktikum Biophysik\E8\Results 2.csv")
```

```python
data.head()
```

```python
angles = np.array(data['Angle'])
```

```python
seaborn.kdeplot(angles,color='red')
plt.hist(angles,facecolor='green',alpha=0.75,density=True)

(mu, sigma) = norm.fit(angles)

plt.xlabel('Angle')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ (no\ specimen)\:}\ \mu=%.3f,\ \sigma=%.3f$' %(mu, sigma))
plt.grid(True)
```

```python
f = plt.gca()
f = seaborn.kdeplot(angles,color='red')

lines = seaborn.kdeplot(angles,color='red').get_lines()[0].get_data()
```

```python
plt.plot(lines[0],[-math.log(x) for x in lines[1]],color='r')
plt.title('Free Energy Landscape (no specimen)')
plt.xlabel('Angle')
plt.ylabel('Free Energy')
```
